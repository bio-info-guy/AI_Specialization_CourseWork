```python
# Reusing some functions from previous assignments for preprocessing
from sklearn.preprocessing import LabelEncoder
# a class for preprocessing panda dataframes that has categorical values 
# into numeric dataframes that can later be converted back into original pandas dataframe
class LabelEncoderWrapper():    
    def __init__(self, df):
        self.df = df
        self.dict={}
    
    def convert2num(self):
        type_name=zip(self.df.dtypes, self.df.columns.values)
        df_cp=self.df.copy()
        for (dtype, name) in type_name:
            if str(dtype) == "object" or str(dtype) == "bool":
                le, df_cp[name] = self.cat2num(df_cp[name])
                self.dict[name] = le
        return df_cp
    
    def cat2num(self, cat):
        le=LabelEncoder()
        num=le.fit_transform(cat)
        
        return le, num     
        
    def original(self):
        return self.df
    
    def num2cat(self, name, num):
        assert name in self.dict.keys(), "No key value for "+name+" was found in the dictionary"
        le=self.dict[name]
        return le.inverse_transform(num)
```

```python
def OneHot(T, reverse=False):
    # transform target vector into a oneHot encoding 2d matrix
    if reverse:
        assert len(T.shape) == 2, 'need 2d matrix fro onehot encoding'
        return np.argmax(T, axis=1)
    targetType=list(set(T.flatten()))
    oneHotT=np.zeros((T.shape[0], len(targetType)))
    for i in range(T.shape[0]):
        oneHotT[i,int(T[i])-1]=1
    return oneHotT
```

```python
import abc
from abc import ABC, abstractmethod
# Super class for machine learning models 

class BaseModel(ABC):
    """ Super class for ITCS Machine Learning Class"""
    
    @abstractmethod
    def train(self, X, T):
        pass

    @abstractmethod
    def use(self, X):
        pass

class LinearModel(BaseModel):
    """
        Abstract class for a linear model 
        
        Attributes
        ==========
        w       ndarray
                weight vector/matrix
    """

    def __init__(self):
        """
            weight vector w is initialized as None
        """
        self.w = None
        
    def reset(self):
        pass
    
    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abstractmethod
    def train(self, X, T):
        """
            train linear model
            
            parameters
            -----------
            X     2d array
                  input data
            T     2d array
                  target labels
        """        
        pass
    
    @abstractmethod
    def use(self, X):
        """
            apply the learned model to input X
            
            parameters
            ----------
            X     2d array
                  input data
            
        """        
        pass 

class Classifier(BaseModel):
    """
        Abstract class for classification 
        
        Attributes
        ==========
        meanX       ndarray
                    mean of inputs (from standardization)
        stdX        ndarray
                    standard deviation of inputs (standardization)
    """

    def __init__(self, ):
        self.meanX = None
        self.stdX = None
        self.w=None

    def normalize(self, X):
        """ standardize the input X """
        
        if not isinstance(X, np.ndarray):
            X = np.asanyarray(X)

        if np.any(self.meanX) == None or np.any(self.stdX) == None:
            self.meanX = np.mean(X, 0)
            self.stdX = np.std(X, 0)

        # TODO: Finish this normalization
        Xs = (X-self.meanX)/self.stdX
        return Xs

    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abc.abstractmethod
    def train(self, X, T):
        pass
    
    @abc.abstractmethod
    def use(self, X):
        pass 
```

### Misc Functions for Confusion Matrix, ROC-AUC curves and other evaluation metrics

```python
from sklearn.metrics import roc_curve, auc
def roc_auc( T, Y):
    fpr, tpr, _ =roc_curve( np.argmax(T, axis=1), Y)
    roc_auc=auc(fpr, tpr)
    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='darkorange',
    lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show() # Code from sklearn tutorial for plotting of auc roc curves

def confusion_matrix(Y, T):
    if len(Y) != len(T):
        raise ValueError("Wrong prediction and target length!")
    
    classes = np.unique(T)
    n_classes = len(classes)
    
    cfm = pd.DataFrame(np.zeros((n_classes, n_classes)), index=classes, columns=classes, dtype=int)
    
    Tidx = [T == c for c in classes]
    for c in classes:
        pred_idx = Y == c
        cfm.loc[c, :] = [np.sum(np.logical_and(pred_idx, tidx)) for tidx in Tidx]
    
    return cfm

def plot_confusion_matrix(cm, classes=["defense", "attack"],
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.xlabel('True label')
    plt.ylabel('Predicted label')

def evaluation_metrics(Y, T):
    cfm=confusion_matrix(Y=Y, T=T).values
    fpr=cfm[1,0]/(cfm[0,0]+cfm[1,0])
    precision=cfm[1,1]/(cfm[1,1]+cfm[1,0])
    recall=cfm[1,1]/(cfm[1,1]+cfm[0,1])
    specificity=cfm[0,1]/(cfm[0,1]+cfm[1,0])
    plot_confusion_matrix(cfm)
    F1=2*precision*recall/(precision+recall)
    MCC=(cfm[1,1]*cfm[0,0]-cfm[1,0]*cfm[0,1])/np.sqrt((cfm[1,1]+cfm[1,0])*(cfm[1,1]+cfm[0,1])*(cfm[0,1]+cfm[1,0])*(cfm[0,1]+cfm[0,0]))
    print("Precision:\t" + str(precision) )
    print("Recall:\t" + str(recall) )
    print("Specificity:\t" + str(specificity) )
    print("FPR:\t" + str(fpr) +"\n")
    print("F1 Score:\t" + str(F1) )
    print("MCC Score:\t" + str(MCC) )
```

### Logistic Regression model for classification from assignment 2

```python
from mpl_toolkits.mplot3d import Axes3D

class LogisticRegression(Classifier):
    
    def __init__(self, **params):
        Classifier.__init__(self)
        self.rho=params.pop("rho", 0.1)
        self.niter=params.pop("niter", 30)
        self.binary=params.pop("binary", False)
        self.error=[]
        self.wp=None
        
    def rmse(self, T, Y): 
        return np.sqrt(np.mean((T-Y)**2))
    
    def compare(self, X, T, w, wp):
        y = self.softmax(X,w)[:,1]
        yp = self.softmax(X,wp)[:,1]
        return 1 if np.sum((y - T)**2) <= np.sum((yp - T)**2) else -1
    
    def softmax(self, X, w):
        return np.exp(X@w-np.max(X@w, axis=1).reshape((-1,1)))/np.sum(np.exp(X@w-np.max(X@w, axis=1).reshape((-1,1))),axis=1)[:,np.newaxis]
    
    def train(self, X, T, Xval=None, Tval=None, **params): # T is onehot vector
        oneHotT=T
        if len(T.shape) == 1:
            oneHotT=OneHot(T)
        else:
            T=OneHot(T, reverse=True)
        X=self.normalize(X)
        X1=self.add_ones(X)
        self.w=np.zeros((X1.shape[1], oneHotT.shape[1]))
        self.wp=copy(self.w)
        pocket=params.pop('pocket', False)
        acc=[]
        val_err=[]
        val_acc=[]
        for step in range(self.niter):
            if pocket:
                for i in range(X1.shape[0]):
                    self.w=self.wp+self.rho*X1[i].reshape((1,-1)).T@(oneHotT[i]-self.softmax(X1[i].reshape((1,-1)),self.wp))
                    if self.compare(X1,T,self.w, self.wp) == 1:
                        self.wp=copy(self.w)
            else:            
                self.w=self.w+self.rho*X1.T@(oneHotT-self.softmax(X1,self.w))
                self.wp=self.w
            self.error.append(self.use(X, oneHotT)['error'])
            acc.append(self.use(X, oneHotT)['accuracy'])
            if np.any(Xval) and np.any(Tval):
                val_err.append(self.use(Xval, Tval)['error'])
                val_acc.append(self.use(Xval, Tval)['accuracy'])
        return {'train_error':self.error, 'train_acc':acc, 'val_error': val_err, 'val_acc': val_acc}
            
    def use(self, X, T=None, threshold=0.5, probs=False):
        assert np.any(self.wp) != None, "Weights not learned yet"
        X = (X - self.meanX) / self.stdX
        X1=self.add_ones(X)
        Y=self.softmax(X1, self.wp)
        r=np.argmax(Y, axis=1)
        acc=None
        if self.binary:
            if probs:
                return Y[:,1]
            return np.array([1 if y[1] >= threshold else 0 for y in Y])
        if np.any(T):
            error=np.mean(self.rmse(T, Y))
            acc= np.sum(r == OneHot(T, reverse=True))/float(len(r))
        return {'pred':Y, 'class':r, 'error':error, 'accuracy':acc}
    
    def errorplot(self):
        assert self.error != [], "no training was performed"
        error=np.hstack((np.array(range(len(self.error))), np.array(self.error)))
        plt.plot(np.array(self.error))
        
    def roc_auc(self, X, T):
        fpr, tpr, _ =roc_curve( T, self.use(X, threshold=0.5, probs=True))
        plot_roc(fpr, tpr)
        
```

### Cross Validation Functions for Model Selection

* The ** partition **function partitions the data into number of parts denote by K 

*  The **splitData **functions splits the data into the kth parition and the rest of the data, input is result given by partition function 

*  The** parseParam **function parses for parameters that need are to be tested and returns the name of those parameters, their values and a matrix to store the result for different combination of the parameters 

* The ** CV **function performs cross-validation on the Training data and eventually returns an dictionary object that includes the best model ('model'), parameters used to train that model ('best_params'), the model's training error and accuracy trace ('trace'), the matrix of different parameter combination training errors ('pMatMean'), every training validation trace in the process in an ordered list ('training_trace') and most importantly the results on the test data in a dictionary ('test') .  **(parentheses are the keywords to access the items)**

```python
def partition(x, K):
    ptsize=x.shape[0]//K
    start=0
    parts=[]
    for i in range(K):
        if i == K-1:
            parts.append(x[start:,])
        else:
            parts.append(x[start:start+ptsize,])
            start += ptsize
    return parts

def splitData(parts, k):
    x=np.array([])
    for i in range(len(parts)):
        if i == k:
            continue
        else:
            if np.any(x):
                x=np.vstack((x, parts[i]))
            else:
                x=parts[i]
    return parts[k], x

def parseParams(params):
    Params_test={}
    p_name=[]
    p_vals=[]
    all_paramN=list(params.keys())
    for p in all_paramN:
        if type(params[p]) == tuple:
            #Params_test[p]=params[p]
            p_vals.append(params[p])
            p_name.append(p)
            del params[p]
    dimensions=[len(p) for p in p_vals ]
    paramsMat = np.zeros(tuple(dimensions))
    return params, p_name, p_vals, paramsMat

def CV(X, T, Xtest, Ttest, K, Model, **params):
    assert type(Xtest) in (list, tuple) and type(Ttest) in (list, tuple) and len(Xtest) == len(Ttest),'provide the testing data as list or tuple'
    Xparts=partition(X, K)
    Tparts=partition(T, K)
    params, p_name, p_vals, pMat = parseParams(params)
    Params={}
    Kfold_pMat=[]
    train_trace=[]
    for pt in range(K):
        print('validating on '+str(pt))
        xval, xtrain = splitData(Xparts, pt)
        tval, ttrain = splitData(Tparts, pt)
        for ind in np.ndindex(pMat.shape):
            newparams=copy(params)
            for i in range(len(ind)):
                newparams[p_name[i]]=p_vals[i][ind[i]]
            model=Model(**newparams)
            train_trace.append(model.train(xtrain, ttrain, xval, tval, **newparams))
            result=model.use(xval, tval)
            pMat[ind]=result['error']
        Kfold_pMat.append(pMat)
    pMatMean=np.mean(np.array(Kfold_pMat), axis=0)
    pMatStd=np.std(np.array(Kfold_pMat), axis=0)
    isnan=np.isnan(pMatMean)
    pMatMean[isnan]=np.inf
    all_inds = np.argwhere(pMatMean == np.min(pMatMean))
    Params['best_params']=[]
    Params['test'] = []
    Params['train_trace']=train_trace
    for ind in all_inds:
        best_param={}
        for i in range(len(ind)):
            best_param[p_name[i]]=p_vals[i][ind[i]]
            params[p_name[i]]=p_vals[i][ind[i]]
        Params['best_params'].append(best_param)
    Params['pMatMean']=pMatMean
    Params['pMatStd']=pMatStd
    model=Model(**params)
    trace=model.train(X, T, **params)
    Params['model']=model
    Params['trace']=trace
    for xtest, ttest in zip(Xtest, Ttest):
        Params['test'].append(model.use(xtest, ttest))
    return Params
                
                    
    
```

```python
# standardization class
class Standardizer: 
    """ class version of standardization """
    def __init__(self, X, explore=False):
        self._mu = np.mean(X,0)  
        self._sigma = np.std(X,0)
        if explore:
            print ("mean: ", self._mu)
            print ("sigma: ", self._sigma)
            print ("min: ", np.min(X,0))
            print ("max: ", np.max(X,0))

    def set_sigma(self, s):
        self._sigma[:] = s

    def standardize(self,X):
        return (X - self._mu) / self._sigma 

    def unstandardize(self,X):
        return (X * self._sigma) + self._mu 
```

```python
#from grad import scg, steepest
import numpy as np
from matplotlib import pyplot as plt
from copy import copy
from grad import scg

""" Custom Version of Neural Network:
        Mini-batch training
        Optimization:
            1. scaled conjugate gradient (scg) 
            2. stochastic gradient descent (sgd)
            3. steepest descent (this is not converging and is same as sgd)
        Activation functions:
            1. logistic sigmoid (logi)
            2. hyperbolic tangent (tanh)
            3. leaky rectified linear unit (relu)
        Cost functions:
            1.mean squared error (mse)
            2.cross entropy (cre)

    example usage:
        X = numpy.array([0,0,1,0,0,1,1,1]).reshape(4,2)
        T = numpy.array([0,1,1,0,1,0,0,1]).reshape(4,2)

        nn = nnet.NeuralNet([2,3,2])
        nn.train(X,T, wprecision=1e-20, fprecision=1e-2)
        Y = nn.use(X)

    The model can be set to do:
        classification: target vector has to be a oneHot encoding
            1. automatically uses cross entropy with softmax for non-linear logistic regression
        regression: one dimensional vector
            2. must use least mean square
"""


class NeuralNet:
    """ neural network class for regression or classification
        
        Parameters
        ----------
        nunits: list
            the number of inputs, hidden units, and outputs

        Methods
        -------
        set_hunit    
            update/initiate weights

        pack 
            pack multiple weights of each layer into one vector

        forward
            forward processing of neural network

        backward
            back-propagation of neural network

        train
            train the neural network

        use
            appply the trained network for prediction

        Attributes
        ----------
        Nlayers
            the number of hidden unit layers 

        rho
            learning rate

        Wlayer
            list representation of W
            
        W
            weights in one dimension (Wlayer is referencing W)

        stdX
            standardization class for data
            
        stdT
            standardization class for target

        Notes
        -----
        
    """
    
    def __init__(self, **params):
        self.hiddenN=params.pop("hiddenN", [2,4,3])
        self.batchsize=params.pop("batchsize", 1)
        self.cost_f=params.pop('cost_function', 'mse')
        self.actv_f=params.pop('activation_function', 'tanh')
        self.Z=None
        self.Zgrad=None
        self.A=None
        self.W=None
        self.grad=None
        self.Wlayer=None
        self.error=None
        self.epochs=0
        self.batchnum=0
        self.classification=params.pop('classification', False)
        self.__initialize__=False
        
        
    def add_ones(self, x):
        return np.insert(arr=x, axis=1,obj=0,values=1)
    
    def pack(self, w):
        return np.hstack(map(np.ravel, w))
    
    def unpack(self, weights):
        self.W[:] = weights[:]
        
    def cp_weight(self):
        return copy(self.W)
        
    def generate_batch(self):
        assert self.batchsize != None, 'batchsize was not given'
        nextn=self.batchnum + self.batchsize
        N=self.N
        if nextn > N:
            self.epochs +=1
            Xbatch=np.vstack((self.X[self.batchnum:], self.X[:nextn-N]))
            Tbatch=np.vstack((self.T[self.batchnum:], self.T[:nextn-N]))
            self.batchnum=nextn-N
        else:
            Xbatch=self.X[self.batchnum:nextn]
            Tbatch=self.T[self.batchnum:nextn]
            self.batchnum=nextn
        return Xbatch, Tbatch
        
    def noactiv_f(self, x, gradient=False):
        if gradient:
            return 1
        else:
            return x
        
    def tanh(self, x, gradient=False):
        if gradient:
            return 1-np.tanh(x)**2
        else:
            return np.tanh(x)
        
    def logi(self, x, gradient=False):
        if gradient:
            return (1/(1+np.exp(-x)))*(1-(1/(1+np.exp(-x))))
        else:
            return 1/(1+np.exp(-x))
    
    def relu(self, x, gradient=False):
        a=x > 0
        a=a.astype(np.float32)
        active=(1-a)*self.alpha+a
        if gradient:
            return active
        else:
            return active*x
    
    def softmax_grad(self, s):
    # from Aaron: https://stackoverflow.com/questions/33541930/how-to-implement-the-softmax-derivative-independently-from-any-loss-function
    # This was hard to implement, since for each sample there is a jacobian matrix since partial derivatives are not zero in the softmax
    # input s is softmax value of the original input x. Its shape is (1,n) 
    # e.i. s = np.array([0.3,0.7]), x = np.array([0,1])

    # make the matrix size  n^2.
        jacobian_m = np.diag(s)

        for i in range(len(jacobian_m)):
            for j in range(len(jacobian_m)):
                if i == j:
                    jacobian_m[i][j] = s[i] * (1-s[i])
                else: 
                    jacobian_m[i][j] = -s[i]*s[j]
        return jacobian_m
    
    def softmax(self, x, gradient=False):
        Xexp=np.exp(x)
        Y=Xexp/np.sum(Xexp, axis=1).reshape((-1,1))
        if gradient:
            G=np.array([self.softmax_grad(s) for s in Y])
            #if self.classification:
                #return 1
            return G
        else:
            return Y
    
    def mse(self, Y, T, gradient=False):
        if gradient:
            return Y-T
        else:
            return (T-Y)**2
        
    def cre(self, Y, T, gradient=False):
        if gradient:
            #if self.classification:
                #return Y-T # This was not modulized, Y-T is product of activation function gradient and the cross entropy gradient
            Y[np.where(Y==0)]=1e-8
            return -T/Y # this is modulized
        else:
            Y[np.where(Y==0)]=1e-8
            Y[np.where(Y==1)]=0.9999999
            return -T*np.log(Y)
        
            
    def set_hunit(self):
        self.Wlayer=[]
        total_params=np.sum([(self.hiddenN[i]+1)*self.hiddenN[i+1] for i in range(self.Nlayers-1)])
        self.W=np.random.uniform(-0.1, 0.1, total_params)
        start=0
        for layer in range(self.Nlayers-1):
            self.Wlayer.append(self.W[start:(start+(self.hiddenN[layer]+1)*self.hiddenN[layer+1])].reshape((self.hiddenN[layer]+1,self.hiddenN[layer+1])))
            start +=(self.hiddenN[layer]+1)*self.hiddenN[layer+1]
        
    def set_actvf(self, alpha=0.1):
        activf={"tanh":self.tanh, "logi": self.logi, "relu":self.relu} 
        if type(self.actv_f) == str:
            assert self.actv_f in activf.keys(), 'does not support this activation function'
            self.actv_f=[activf[self.actv_f]]*(self.Nlayers-1)
        elif type(self.actv_f) in [list, tuple]:
            assert len(self.actv_f) == self.Nlayers, 'not enough activation functions were given, please give in order as well'
            self.actv_f = [activf[a] for a in self.actv_f]
           
        if self.classification:
            self.actv_f[-1]=self.softmax
        else:
            self.actv_f[-1]=self.noactiv_f

            
    def set_costf(self):
        if self.cost_f == "mse" or not self.classification:
            self.cost_f = self.mse
        elif self.cost_f == "cre":
            self.cost_f = self.cre
        
    def forward(self, X, predict=False):
        assert np.any(self.W) and self.Wlayer , 'Weights have not been initialized yet'
        self.Z=[0 for i in range(self.Nlayers) ]
        self.Zgrad=[0 for i in range(self.Nlayers-1)]
        self.A=[0 for i in range(self.Nlayers) ]
        self.A[0]=X
        
        for i in range(self.Nlayers-2):
            self.Z[i+1]=self.A[i]@self.Wlayer[i]
            self.A[i+1]=self.add_ones(self.actv_f[i](self.Z[i+1]))
        self.Z[-1]=self.A[-2]@self.Wlayer[-1]
        self.A[-1]=self.actv_f[-1](self.Z[-1])
        if predict:
            return self.A[-1]
        
    def backward(self, T):
        self.error=np.sum(self.cost_f(self.A[-1], T))
        if self.classification or self.actv_f[-1] == self.softmax:
            self.Zgrad[-1]=self.cost_f(self.A[-1], T, gradient=True)[:,np.newaxis,:]*self.actv_f[-1](self.Z[-1], gradient=True)
            self.Zgrad[-1]=np.sum(self.Zgrad[-1], axis=2)
        else:
            self.Zgrad[-1]=self.cost_f(self.A[-1], T, gradient=True)*self.actv_f[-1](self.Z[-1], gradient=True)
        
        for i in np.arange(len(self.Wlayer)-1)+1:
            self.Zgrad[-(i+1)]=(self.Zgrad[-i]@self.Wlayer[-i][1:].T)*self.actv_f[-(i+1)](self.Z[-(i+1)], gradient=True)
        
            
    def gradientf(self, weights):
        self.unpack(weights)
        self.forward(self.Xb)
        self.backward(self.Tb)
        grd=[]
        rh=self.rho/self.batchsize
        for i in np.arange(len(self.Wlayer))+1:
            if i==len(self.Wlayer):
                lmbterm = 0
            else:
                lmbterm = self.lbd * np.vstack((np.zeros((1, self.Wlayer[-i].shape[1])),
                            self.Wlayer[-i][1:,]))
            grd.insert(0, (rh*(self.A[-(i+1)].T@self.Zgrad[-i])+lmbterm))
        return self.pack(grd)
    
    def optimtargetf(self, weights):
        self.unpack(weights)
        self.Xb, self.Tb = self.generate_batch()
        self.forward(self.Xb)
        Wnb=np.array([])
        for i in range(self.Nlayers-1):
            if len(Wnb)==0: 
                Wnb=self.Wlayer[i][1:,].reshape(self.Wlayer[i].size-self.Wlayer[i][0,].size,1)
            else: 
                Wnb = np.vstack((Wnb,self.Wlayer[i][1:,].reshape(self.Wlayer[i].size-self.Wlayer[i][0,].size,1)))
            wpenalty = self.lbd * np.dot(Wnb.flat ,Wnb.flat)
        return np.sum(self.cost_f(self.A[-1], self.Tb))+wpenalty
    
    def weight_update(self):
        self.Xb, self.Tb = self.generate_batch()
        self.W[:] = self.W - self.gradientf(self.cp_weight())

    def init_network(self, X, T):
        self.hiddenN.insert(0, X.shape[1])
        self.hiddenN.append(T.shape[1])
        self.Nlayers=len(self.hiddenN) # total number of layers equal: 1 (input_layer) + # of hidden layers + 1 (output_layer)
        self.set_hunit()
        self.set_actvf()
        self.set_costf()
        self.__initialize__=True
    
    def train(self, X, T, Xval=None, Tval=None, **params):
        self.alpha=params.pop('alpha', 0.1)
        self.lbd=params.pop('lbd', 0.01) 
        self.rho=params.pop('rho', 0.1)
        self.batchsize=params.pop('batchsize', X.shape[0])
        self.N=X.shape[0]
        epochs=params.pop('epochs', 100)
        wprecision=params.pop('wprecision',1e-20)
        fprecision=params.pop('fprecision',1e-20)
        self.stX=params.pop('standardX', True)
        self.stT=params.pop('standardT', False)
        optim = params.pop('optim', 'sgd')
        nIter=epochs*int(self.N/self.batchsize)
        wtracep = params.pop('wtracep', True)
        ftracep = params.pop('ftracep', True)
        Xs=X
        
        if self.stT:
            self.standardT=Standardizer(T)
            T=self.standardT.standardize(T)
        if self.stX:
            self.standardX=Standardizer(X)
            Xs=self.standardX.standardize(X)
            
        X1=self.add_ones(Xs)
        self.X=X1
        self.T=T
        self.init_network(X,T)
        
        if optim == 'sgd':
            trn_err=[]
            trn_acc=[]
            val_err=[]
            val_acc=[]
            niter=-1
            while self.epochs < epochs:
                self.weight_update()
                if self.epochs - niter == 1:
                    train_r=self.use(X, T)
                    trn_err.append(train_r['error'])
                    trn_acc.append(train_r['accuracy'])
                    niter=self.epochs
                    if np.any(Xval) and np.any(Tval):
                        assert Xval.shape[0] == Tval.shape[0], 'not matching validation data'
                        val_r=self.use(Xval, Tval)
                        val_err.append(val_r['error'])
                        val_acc.append(val_r['accuracy'])
            return {'train_error':trn_err, 'train_acc':trn_acc, 'val_error': val_err, 'val_acc': val_acc}
                
        elif optim == 'scg': # using scaled conjugate gradient
            result = scg(self.cp_weight(), self.gradientf, self.optimtargetf,
                                        wPrecision=wprecision, fPrecision=fprecision, 
                                        nIterations=nIter,
                                        wtracep=wtracep, ftracep=ftracep,
                                        verbose=False)
            self.unpack(result['w'][:])
            self.f = result['f']
            return result
        
        elif optim == 'steepest':# for unknown reasons, this function will not converge unlike the sgd function above
            result = steepest(self.cp_weight(), self.gradientf, self.optimtargetf,
                                nIterations=epochs*int(self.N/self.batchsize),
                                xPrecision=wprecision, fPrecision=fprecision,
                                xtracep=wtracep, ftracep=ftracep )
            self.unpack(result['w'][:])
            return result
    
    def use(self, X, T=None):
        assert self.__initialize__ == True, 'Network has not been initialized, likely Network has not been trained'
        Xs=X
        if self.stX:
            Xs=self.standardX.standardize(X)
            
        X1=self.add_ones(Xs)
        pred=self.forward(X1, predict=True)
        r=np.argmax(pred, axis=1)
        if self.stT:
            pred=self.standardT.unstandardize(pred)
            r=None # r is actual result of classification prediction
        error=None
        acc=None
        if np.any(T):
            error=np.mean(self.cost_f(pred, T))
            if self.classification: # accuracy of classification, regression doesn't have accuracy
                acc = np.sum(r == OneHot(T, reverse=True))/float(len(r))
        return {'pred':pred, 'class':r, 'error':error, 'accuracy':acc}
    
    
    
```

### LMS model for regression from Assignment 1

```python
import collections # for checking iterable instance

# LMS class 
class LMS(LinearModel):
    """
        Lease Mean Squares. online learning algorithm
    
        attributes
        ==========
        w        nd.array
                 weight matrix
        alpha    float
                 learning rate
    """
    def __init__(self, **params):
        LinearModel.__init__(self)
        self.rho = params.pop('rho', 0.1)
        self.bias_added=params.pop('bias_added', False)
    def reset(self,alpha, bias_added=False):
        self.w=None
        self.alpha=alpha
        self.bias_added=bias_added
    # batch training by using train_step function
    def train(self, X, T):
        N=X.shape[0]
        Niter=params.pop("niter", 30)
        for i in Niter:
            for k in range(N):
                self.train_step(X[k,:],T[k,:])
            
    # train LMS model one step 
    # here the x is 1d vector
    def train_step(self, x, t):
        if self.bias_added==False:
            x=np.hstack((1,x))
        if np.any(self.w) == None:
            self.w=np.zeros(x.shape[0])
        self.w = self.w - self.alpha*(np.dot(x,self.w)-t)*x
    # apply the current model to data X
    def use(self, X):
        assert self.w.any() != None, "weights have not been learned"
        if self.bias_added==False:
            X1 = self.add_ones(X)
        return np.dot(X1,self.w).reshape((-1,1))
        
```
